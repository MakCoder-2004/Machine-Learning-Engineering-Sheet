# TensorFlow Fundamentals

## Creating tensors using `tf.constant()`

```python
vector = tf.constant(value, dtype=None, shape=None, name='Const')
```

```python
# Creating matrix
matrix = tf.constant([[10, 7],
                      [7, 10],
                      [10,7]])
print(matrix)
print("ndim =", matrix.ndim)
print("shape =", matrix.shape)
```

```python
# Changing tensor [We cannot reassign in constants "It will give an error"]
unchangeable_tensor = tf.constant([10,7])
unchangeable_tensor[0].assign(8)
unchangeable_tensor
```

## Creating tensors using `tf.variable()`

```python
changeable_tensor = tf.constant([10,7])
print(unchangeable_tensor)
```

```python
# Changing tensor
changeable_tensor = tf.Variable([10,7])
changeable_tensor[0].assign(8)
changeable_tensor
```

## Indexing and expanding tensors

```python
tensor2 = tf.constant([[10,7],[4,6]])

# Get the last element of each row
tensor2[:,-1]
```

```python
tensor3 = tf.constant([[1,2,3], [4,5,6], [7,8,9]])

# Insert another dimension
print(tensor3[tf.newaxis, :, :]) # => [[[1,2,3], [4,5,6], [7,8,9]]]
print(tensor3[:, tf.newaxis, :]) # => [[[1,2,3]], [[4,5,6]], [[7,8,9]]]
print(tensor3[:, :, tf.newaxis]) # => [[[1],[2],[3]], [[4],[5],[6]],[[7],[8],[9]]]
```

## Reshaping Matrix

```python
tf1 = tf.constant([[10, 7],
                   [3, 4],
                   [1, 2]])  # shape = (3, 2)
```

```python
# Reshape with specific dimensions
reshaped = tf.reshape(tf1, shape=(2, 3))
print("original shape:", tf1.shape)
print("reshaped shape:", reshaped.shape)
print(reshaped)
```

```python
# reshape with flipping the dimensions
reshaped = tf.transpose(tf1)
print("original shape:", tf1.shape)
print("reshaped shape:", reshaped.shape)
print(reshaped)
```

```python
# Difference between transpose and reshape
tensor = tf.constant([[10, 7],
                      [3, 4],
                      [1, 2]])  # shape = (3, 2)

print("original shape:", tensor.shape)
print("original:\n", tensor.numpy())

reshaped = tf.reshape(tensor, (2, 3))
print("reshaped shape:", reshaped.shape)
print("reshaped:\n", reshaped.numpy())

transposed = tf.transpose(tensor)
print("transposed shape:", transposed.shape)
print("transposed:\n", transposed.numpy())
```

## Tensor Aggregation

```python
tensor = tf.constant(np.random.randint(-5, 100, size=50))
tensor
```

```python
# abs
abs = tf.abs(tensor)
print(abs)
```

```python
# minimum
minimum = tf.reduce_min(tensor)
print(minimum)
```

```python
# maximum
maximum = tf.reduce_max(tensor)
print(maximum)
```

```python
# sum
sum = tf.reduce_sum(tensor)
print(sum)
```

```python
# mean
mean = tf.reduce_mean(tensor)
print(mean)
```

```python
# variance
var = tf.math.reduce_variance(tf.cast(tensor, dtype=tf.float32))
print(var)
```

```python
# standard deviation
std = tf.math.reduce_std(tf.cast(tensor, dtype=tf.float32))
print(std)
```

## Check TensorFlow Running on GPU or CPU

```python
import tensorflow as tf
import os

# Suppress TensorFlow warnings (optional)
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'  # 0=all, 1=filter INFO, 2=filter WARNING, 3=filter ERROR

# Check TensorFlow version
print(f"TensorFlow Version: {tf.__version__}")

# Check GPU availability
print("\n" + "="*80)
print("GPU Configuration")
print("="*80)

# List physical devices
physical_devices = tf.config.list_physical_devices()
print(f"\nAll Physical Devices: {len(physical_devices)}")
for device in physical_devices:
    print(f"  - {device}")

# List GPU devices
gpus = tf.config.list_physical_devices('GPU')
print(f"\nGPU Devices Found: {len(gpus)}")
for i, gpu in enumerate(gpus):
    print(f"  GPU {i}: {gpu}")
    # Get GPU details
    gpu_details = tf.config.experimental.get_device_details(gpu)
    if gpu_details:
        print(f"    Details: {gpu_details}")

# Check if GPU is available
if gpus:
    print("\n✓ GPU is available and will be used for training!")
    try:
        # Set memory growth to prevent TensorFlow from allocating all GPU memory at once
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print("✓ Memory growth enabled for GPU(s)")
    except RuntimeError as e:
        print(f"Memory growth setting error: {e}")
else:
    print("\n✗ No GPU found. Training will use CPU.")

# Verify CUDA and cuDNN
print(f"\nBuilt with CUDA: {tf.test.is_built_with_cuda()}")
print(f"GPU Available: {tf.test.is_gpu_available(cuda_only=False, min_cuda_compute_capability=None) if hasattr(tf.test, 'is_gpu_available') else 'N/A (deprecated)'}")

print("="*80 + "\n")
```
